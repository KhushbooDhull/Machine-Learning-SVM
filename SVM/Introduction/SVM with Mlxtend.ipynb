```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
x=iris.data[:,:2] #independant features- sepal length, sepal width
y=iris.target #target variable
```

```python
print(x.shape)
print(y.shape)
```

```python
print(type(x))
print(type(y))
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
from sklearn.svm import SVC,SVR
```

### Plot SVM with different Kernals

```python
kernel=['linear','rbf','poly']
for i in kernel:
    model=SVC(kernel=i) #default C=1.0
    model.fit(x,y)
    plot_decision_regions(x,y,clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with Kernel {i}' )
    plt.show()
```

### Linear Kernel with Different C

```python
c=[0.0001,0.001,0.01,0.1,1,10,100]
for i in c:
    model=SVC(kernel='linear',C=i) #default C=1.0
    model.fit(x,y)
    plot_decision_regions(x,y,clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with Linear Kernel with c= {i}' )
    plt.show()
```

### Poly Kernel with Different C

```python
c=[0.0001,0.001,0.01,0.1,1,10,100]
for i in c:
    model=SVC(kernel='poly',C=i) #default C=1.0
    model.fit(x,y)
    plot_decision_regions(x,y,clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with Poly Kernel with c= {i}' )
    plt.show()
```

### Poly Kernel with Different Degree

```python
deg=[1,2,3,4,5]
for i in deg:
    model=SVC(kernel='poly',degree=i) #default C=1.0
    model.fit(x,y)
    plot_decision_regions(x,y,clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with Poly Kernel with degree= {i}' )
    plt.show()
```

### RBF Kernel with Different Gamma

```python
g=[0.01,0.1,1,10,100]
for i in g:
    model=SVC(kernel='rbf',gamma=i) #default C=1.0
    model.fit(x,y)
    plot_decision_regions(x,y,clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with rbf Kernel with gamma= {i}' )
    plt.show()
```
